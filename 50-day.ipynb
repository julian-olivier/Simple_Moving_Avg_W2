## plotting stock



```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np 

plt.style.use('ggplot')

df = pd.read_csv("./all_stocks_5yr.csv")

nvdia = df[ df['Name'] == 'NVDA'].copy()
nvdia['date'] = pd.to_datetime(nvdia['date'])

fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(nvdia['date'], nvdia['close'], color='#1f77b4', linewidth=2)

ax.set_title('NVIDIA (NVDA) Closing Price', fontsize=16, fontweight='bold')
ax.set_xlabel('Date', fontsize=12)
ax.set_ylabel('Closing Price (USD)', fontsize=12)
plt.xticks(rotation = 45)
plt.tight_layout()

plt.show()
```

## 50-day & 200-day section

```python
if 'date' in nvdia.columns:
   nvdia.set_index('date', inplace=True)

nvdia.sort_index()

start_date = pd.to_datetime('2013-02-08')
start_loc = nvdia.index.get_loc(start_date)

nvdia_50 = nvdia.iloc[start_loc: start_loc + 50]
nvdia_200 = nvdia.iloc[start_loc: start_loc + 200]
```

## SMA function 


```python
def SMA():
    nvdia['SMA_50'] = nvdia['close'].rolling(window=50).mean()
    nvdia['SMA_200'] = nvdia['close'].rolling(window=200).mean()

    display(nvdia)
SMA()
```
