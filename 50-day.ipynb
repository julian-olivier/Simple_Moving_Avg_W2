## plotting stock



```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np 

plt.style.use('ggplot')

df = pd.read_csv("./all_stocks_5yr.csv")

nvdia = df[ df['Name'] == 'NVDA'].copy()
nvdia['date'] = pd.to_datetime(nvdia['date'])

fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(nvdia['date'], nvdia['close'], color='#1f77b4', linewidth=2, label = 'Closing Price')

ax.set_title('NVIDIA (NVDA) Closing Price', fontsize=16, fontweight='bold')
ax.set_xlabel('Date', fontsize=12)
ax.set_ylabel('Closing Price (USD)', fontsize=12)
plt.xticks(rotation = 45)
plt.tight_layout()

plt.show()
```

## SMA function 


```python
def SMA( w:int ):
    data = nvdia['close'].rolling(window=w).mean()

    return data
```

## plot SMA

```python
ax.plot(nvdia['date'], SMA(50), color = "#9eb41f", label = 'SMA50')    # 50 day SMA 
ax.legend()

fig
```
